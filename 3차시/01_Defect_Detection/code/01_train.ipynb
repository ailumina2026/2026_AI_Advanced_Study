## 4. í•™ìŠµ íŒŒë¼ë¯¸í„° ì„¤ì •

```python
# í•™ìŠµ ì„¤ì •
EPOCHS = 30
BATCH_SIZE = 8
IMAGE_SIZE = 640
MODEL_SIZE = 'yolo11n.pt'  # yolo11n.pt, yolo11s.pt, yolo11m.pt, yolo11l.pt, yolo11x.pt

print("=== í•™ìŠµ ì„¤ì • ===")
print(f"Epochs: {EPOCHS}")
print(f"Batch Size: {BATCH_SIZE}")
print(f"Image Size: {IMAGE_SIZE}")
print(f"Model: {MODEL_SIZE}")
```

## 5. YOLOv11 ëª¨ë¸ í•™ìŠµ

ì‚¬ì „ í•™ìŠµëœ YOLOv11 ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ì „ì´í•™ìŠµ(Transfer Learning)ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.

```python
# ì‚¬ì „ í›ˆë ¨ëœ YOLOv11 Detection ëª¨ë¸ ë¡œë“œ
model = YOLO(MODEL_SIZE)
print(f"âœ… {MODEL_SIZE} ëª¨ë¸ ë¡œë“œ ì™„ë£Œ")
```

```python
# í•™ìŠµ ì‹¤í–‰
results = model.train(
    data='data/config.yaml',
    epochs=EPOCHS,
    batch=BATCH_SIZE,
    imgsz=IMAGE_SIZE,
    project='runs',  # í”„ë¡œì íŠ¸ í´ë”
    name='defect',  # ê³ ì •ëœ í´ë”ëª…
    exist_ok=True,  # ê¸°ì¡´ í´ë” ë®ì–´ì“°ê¸° í—ˆìš©
    val=True,
    verbose=True,
    device=0  # GPU ì‚¬ìš© (CPU: 'cpu')
)

print("\nâœ… í•™ìŠµ ì™„ë£Œ!")
print(f"í•™ìŠµëœ ëª¨ë¸ ì €ì¥ ìœ„ì¹˜: runs/defect/weights/best.pt")
print(f"\nğŸ“ ë‹¤ìŒ ë…¸íŠ¸ë¶ì—ì„œ ì‚¬ìš©í•  ëª¨ë¸ ê²½ë¡œ:")
print(f"   {os.path.join(WORK_DIR, 'runs/defect/weights/best.pt')}")
```
